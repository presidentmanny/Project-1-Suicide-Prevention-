```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt


datefile = "Suicide_Mortality.csv"
```

```python
df = pd.read_csv(datefile,header=1)
df.head()
```

```python
df.YEAR.unique()
```

```python
len(df.STATE.unique())
```

```python
df.dtypes
```

```python
df.DEATHS=df.DEATHS.str.replace(',','')
df.DEATHS=df.DEATHS.astype(int)
```

```python
df.info()
```

```python
var  = df.groupby('YEAR').DEATHS.sum()
#grouped sum of sales at gender level
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('YEAR')
ax1.set_ylabel('Total Death ')
ax1.set_title('Total Death by Year')
var.plot(kind='line')
```

```python
var  = df.groupby('YEAR').RATE.mean()
#grouped sum of sales at gender level
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('YEAR')
ax1.set_ylabel('Average Rate ')
ax1.set_title('Average Rate by Year')
var.plot(kind='line')
```
